```python
#df filtré sur des semaines souhaitées
start_week = 61
end_week = 70

def extract(data_dir: str, prefix: str, start_week: int, end_week: int) -> pd.DataFrame :
    df3 = pd.DataFrame()
    #data_dir = "../batchs"
    for i in range(start_week, end_week+1):
        file_path = os.path.join(data_dir,f'{prefix}_week_{i}.csv')
        if os.path.isfile(file_path):
            batch = pd.read_csv(file_path)
            df3 = pd.concat([df3,batch], sort=True)
    return df3
```

```python
df3 = extract("C:/Users/104658/Vasseur_ML/batchs", "restaurant_1", 122, 126)
df3.head(5)
```

```python
import numpy as np
# Nettoyage des données
def clean(df3: pd.DataFrame) -> pd.DataFrame:
    df3.columns = df3.columns.str.lower().str.replace(' ','_')
    df3['order_date'] = pd.to_datetime(df3['order_date'])
    # CA
    df3['total_product_price'] = df3['quantity']*df3['product_price']
    df3['cash_in'] =  df3.groupby('order_number')['total_product_price'].transform(np.sum)
    df3 = df3.drop(columns=['item_name','quantity','product_price','total_products','total_product_price'])
    df3 = df3.drop_duplicates()
    df3 = df3.reset_index(drop=True)
    return df3
```

```python
df2 = extract("C:/Users/104658/Vasseur_ML/batchs", "restaurant_2", 122, 126)
df2.head(5)
```

```python
# Nettoyage des données
def clean2(df2: pd.DataFrame) -> pd.DataFrame:
    df2 = df2.rename(columns={'order_id':'order_number'})
    df2.columns = df2.columns.str.lower().str.replace(' ','_')
    df2['order_date'] = pd.to_datetime(df2['order_date'])
    # CA
    df2['total_product_price'] = df2['quantity']*df2['product_price']
    df2['cash_in'] =  df2.groupby('order_id')['total_product_price'].transform(np.sum)
    df2 = df2.drop(columns=['item_name','quantity','product_price','total_products','total_product_price'])
    df2 = df2.drop_duplicates()
    df2 = df2.reset_index(drop=True)
    return df2
```

```python
df2 = clean2(df2)
df3 = clean(df3)
```

```python
df3 = df3.resample('1H', on='order_date').sum().reset_index()
```

```python
def merge(df2: pd.DataFrame, df3: pd.DataFrame)->pd.DataFrame:
    df4 = pd.concat([df2,df3], sort=True)
    df4 = df4.drop(columns='order_number')
    df4 = df4.sort_values('order_date')
    df4 = df4.reset_index(drop=True)
    return df4
```

```python
df4 = merge(df2,df3)
df4 = df4.resample('1H', on='order_date').sum().reset_index()
df4.head()
```

```python
def etl(data_dir: str, start_week: int, end_week: int) -> pd.DataFrame :
    df = pd.DataFrame()
    df1 = extract(data_dir,"restaurant_1",start_week,end_week)
    df2 = extract(data_dir,"restaurant_2",start_week,end_week)
    cl = clean(df1)
    cl2 = clean2(df2)
    df = merge(cl,cl2)
    df = df.resample('1H', on='order_date').sum().reset_index()
    return df
```

```python
df = etl("C:/Users/104658/Vasseur_ML/batchs", 122, 126)
df.head(2)
```

```python
# Jour, mois, année, lag
def prepadate(df: pd.DataFrame, order_date: str)-> pd.DataFrame :
    df['Year'] = df[order_date].dt.year 
    df['Month'] = df[order_date].dt.month
    df['Day'] = df[order_date].dt.weekday
    df['Dayname'] = df[order_date].dt.day_name()
    df = pd.get_dummies(df, columns=['Day'], drop_first=True)
    return df
```

```python
df = prepadate(df,'order_date')
df.head()
```

```python
lag_in_week = 1
df = df.set_index('order_date')
df[f'lag_{lag_in_week}W'] = df['cash_in'].shift(7*lag_in_week,'D')
df = df.dropna()
df = df.reset_index()
```
